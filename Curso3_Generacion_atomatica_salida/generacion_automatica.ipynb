# Registro

El registro (**Logging**) en Python es una herramienta esencial que separa a un programador principiante de uno profesional. Mientras que `print()` es como gritar algo al aire en el momento, el **logging** es como llevar un diario detallado, organizado y clasificado de todo lo que ocurre en tu aplicación.

Aquí tienes una explicación estructurada para dominar este tema:

---

## 1. ¿Por qué no usar `print()` en producción?

Aunque `print()` es fácil de usar, tiene grandes limitaciones:

* **Todo o nada:** No puedes apagar los mensajes de `print()` sin borrar el código o usar condiciones complejas.
* **Falta de contexto:** Un `print("Error")` no te dice cuándo ocurrió, en qué archivo o qué tan grave es.
* **Destino rígido:** Por defecto, solo escribe en la consola. Si tu servidor está en la nube, esos mensajes pueden perderse.

---

## 2. Los Niveles de Gravedad (Hierarchy)

El corazón del logging es la capacidad de **priorizar**. Python define 5 niveles estándar:

| Nivel | Cuándo usarlo |
| --- | --- |
| **DEBUG** | Detalles técnicos detallados, útiles solo para diagnosticar problemas. |
| **INFO** | Confirmación de que las cosas funcionan como se espera (ej. "Usuario inició sesión"). |
| **WARNING** | Indicación de que algo inesperado sucedió, pero el programa sigue funcionando. |
| **ERROR** | Un problema grave; el programa no pudo realizar alguna función. |
| **CRITICAL** | Un error fatal; la aplicación podría dejar de funcionar por completo. |

**Dato clave:** Si configuras tu logger en nivel `INFO`, ignorará todos los mensajes `DEBUG`, pero mostrará todos los superiores (`INFO`, `WARNING`, etc.).

---

## 3. Arquitectura del Logging: Los Componentes

Para entender cómo funciona el flujo de un mensaje, imagina una línea de ensamblaje:

1. **Loggers:** Es la interfaz que el programador usa (ej. `logger.info("Hola")`).
2. **Handlers (Manejadores):** Deciden **a dónde** va el mensaje (consola, archivo, email). Un solo logger puede tener varios handlers.
3. **Filters:** Deciden si un mensaje pasa o se descarta basándose en criterios específicos.
4. **Formatters:** Deciden **cómo se ve** el mensaje (añaden la fecha, el nombre del archivo, el nivel, etc.).

---

## 4. El concepto de "Handlers" Avanzados

Como mencionas en tu texto, existen manejadores muy potentes que resuelven problemas del mundo real:

* **RotatingFileHandler:** Evita que tus archivos de log pesen Gigabytes. Cuando el archivo llega a un límite (ej. 10MB), crea uno nuevo y archiva el viejo.
* **SMTPHandler:** Envía un correo electrónico automáticamente si ocurre un error de nivel `CRITICAL`.
* **HTTPHandler:** Envía los logs a un servidor externo o base de datos para análisis centralizado.

---

## 5. Ejemplo de Configuración Profesional

En lugar de usar la configuración básica, un programador suele definir un formato claro para que los errores sean fáciles de rastrear:

```python
import logging

# Crear un logger personalizado
logger = logging.getLogger(__name__)

# Crear un manejador (enviar a un archivo)
file_handler = logging.FileHandler('app_errores.log')
file_handler.setLevel(logging.ERROR) # Solo guardar errores graves en archivo

# Crear un formato profesional
# Formato: Fecha - Nombre - Nivel - Mensaje
formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
file_handler.setFormatter(formatter)

# Añadir el manejador al logger
logger.addHandler(file_handler)

# Uso
logger.error("No se pudo conectar a la base de datos de clientes.")

```

**Resumen:**
El **Logging** te permite observar el comportamiento de tu aplicación en "tiempo real" y de forma histórica, permitiéndote filtrar el ruido (mensajes de depuración) y enfocarte en lo que realmente importa (errores de producción) sin modificar el código constantemente.

# El módulo de registro

El módulo de **logging** en Python es uno de los pilares del desarrollo profesional. A diferencia de un simple mensaje en pantalla, el registro estructurado permite que las aplicaciones "hablen" de lo que sucede en su interior, dejando un rastro que puede ser auditado meses después de que el código se haya ejecutado.

A continuación, detallo los conceptos más importantes para profundizar en tu lectura:

---

### 1. El Flujo de Decisión del Logging

Es útil visualizar el módulo como una serie de filtros. Cuando ejecutas una instrucción como `logging.info("Mensaje")`, el sistema sigue este camino:

1. **Nivel del Logger:** ¿El nivel "INFO" es igual o superior al configurado en el objeto `logger`? Si no, se ignora.
2. **Manejadores (Handlers):** El mensaje se envía a cada destino configurado (consola, archivo, etc.).
3. **Nivel del Handler:** Cada destino tiene su propio filtro. Por ejemplo, puedes querer que la consola muestre todo (**DEBUG**), pero que el archivo solo guarde errores graves (**ERROR**).

---

### 2. Jerarquía y el uso de `__name__`

En el texto mencionas el uso de `logging.getLogger(__name__)`. Esta es la **mejor práctica** en Python por una razón específica:

* **Organización jerárquica:** Si tienes un paquete llamado `mi_app` con un módulo `datos.py`, `__name__` se convertirá en `mi_app.datos`.
* **Control granular:** Esto te permite configurar el registro de manera que puedas silenciar los mensajes de "datos.py" pero mantener activos los de "interfaz.py", todo desde un único archivo de configuración central.

---

### 3. Comparativa de Niveles de Gravedad

Para elegir el nivel correcto, puedes guiarte por esta tabla de intención:

| Nivel | Intención técnica | Ejemplo de uso |
| --- | --- | --- |
| **DEBUG** | Diagnóstico detallado | "Variable `x` tiene el valor 10 en la línea 45" |
| **INFO** | Hitos del proceso | "Cargados 500 registros de la base de datos" |
| **WARNING** | Alerta preventiva | "Memoria al 80%, posible ralentización" |
| **ERROR** | Fallo funcional | "No se pudo enviar el correo, reintentando..." |
| **CRITICAL** | Parada del sistema | "Base de datos principal no encontrada. Abortando." |

---

### 4. Alternativas y el Futuro del Registro

Como señala el texto, existen librerías que llevan el registro al siguiente nivel:

* **Loguru:** Es la alternativa más moderna. Simplifica drásticamente la configuración (no necesitas handlers ni formatters manuales) y añade colores y rotación de archivos de forma nativa.
* **Structlog (Logging Estructurado):** Crucial para aplicaciones modernas (microservicios). En lugar de escribir texto libre, escribe **JSON**.
* *¿Por qué?* Porque las herramientas de análisis de datos (como ElasticSearch o Splunk) pueden leer JSON e indexar campos automáticamente (ej: buscar todos los logs donde `usuario_id == 50`), algo imposible de hacer eficientemente con texto plano.



---

### Resumen de Implementación

Si estás empezando, la configuración más común y recomendada es:

1. **Configurar un Formateador:** Incluye siempre la hora (`%(asctime)s`), el nivel (`%(levelname)s`) y el mensaje.
2. **Usar FileHandler:** Para que los errores persistan aunque la consola se cierre.
3. **No usar `logging.basicConfig` en librerías:** Solo úsalo en el script principal de tu aplicación para evitar conflictos con otros módulos.

# Gestión de excepciones

La **gestión de excepciones** en Python es una técnica fundamental que permite a los desarrolladores manejar errores de forma controlada para evitar que un programa se detenga bruscamente cuando algo sale mal.

A continuación, te explico los conceptos clave basados en el documento:

---

## 1. ¿Qué es una excepción?

Una excepción es un error que ocurre **durante la ejecución** del programa e interrumpe su flujo normal. A diferencia de los errores de sintaxis, que se detectan antes de arrancar el código, las excepciones aparecen cuando el programa ya está funcionando.

**Ejemplos comunes:**

* Intentar dividir un número por cero.


* Intentar abrir un archivo que no existe (**FileNotFoundError**).


* Usar una variable que no ha sido definida (**NameError**).


* Enviar el tipo de dato incorrecto a una función (**TypeError**).



---

## 2. Estructura del manejo de errores: `try`, `except` y `else`

Para gestionar estos fallos, Python utiliza bloques de código específicos:

* 
**`try`**: Aquí colocas el código que *podría* fallar.


* **`except`**: Es el bloque que se ejecuta solo si ocurre un error. Permite notificar al usuario o intentar una recuperación sin que el programa se bloquee.


* 
**`else`**: (Opcional) Se ejecuta únicamente si **no** hubo ninguna excepción en el bloque `try`.



---

## 3. Excepciones vs. Errores de Sintaxis

Es vital entender la diferencia para depurar correctamente:

| Característica | Errores de Sintaxis | Excepciones |
| --- | --- | --- |
| **Cuándo ocurre** | Antes de que el programa se ejecute (fase de lectura).

 | Durante la ejecución del programa.

 |
| **Causa** | Gramática incorrecta, mala sangría o errores tipográficos.

 | Operaciones no válidas o condiciones inesperadas.

 |
| **Solución** | Se debe corregir el código para que pueda arrancar.

 | Se pueden "atrapar" y gestionar con `try/except`.

 |

---

## 4. Filosofía "EAFP" (Es mejor pedir perdón que permiso)

El documento menciona que los programadores expertos de Python (Pythonistas) prefieren evitar el **código defensivo**.

* 
**Código defensivo:** Llenar el programa de condiciones `if` para comprobar si todo es correcto antes de actuar.


* 
**Filosofía Python:** Simplemente intenta realizar la operación y, si falla, maneja la excepción. Esto hace que el código sea más limpio y directo.



**Ejemplo práctico:**
En lugar de preguntar siempre si una clave existe en un diccionario, intentas acceder a ella y usas un `except KeyError` para manejar el caso en que falte.



Claro que sí. Basándome en la información proporcionada, las excepciones son errores que interrumpen el flujo normal de un programa y de los que el código no puede recuperarse automáticamente.

Aquí tienes una explicación detallada de las excepciones más comunes que encontrarás al programar en Python:

---

## Excepciones de Datos y Tipado

Estas ocurren cuando intentas realizar operaciones con datos que no tienen el formato o el tipo esperado.

* 
**ValueError**: Se produce cuando el valor de un parámetro es incorrecto para la operación, aunque su tipo sea técnicamente válido. Por ejemplo, intentar convertir la palabra "hola" a un número entero.


* 
**TypeError**: Ocurre al enviar una cadena cuando una función espera un número entero (`int`), o al llamar a una función con el tipo o número de argumentos incorrectos.


* 
**AttributeError**: Suele deberse a un error tipográfico al intentar llamar a un método o atributo sobre un objeto que no lo posee.



---

## Excepciones de Identificación y Estructura

Estas suelen estar relacionadas con errores en los nombres de los elementos o en la organización del proyecto.

* 
**NameError**: Normalmente sucede debido a un error tipográfico en el nombre de una variable. Es decir, intentas usar una variable que no existe o que no ha sido definida todavía.


* 
**ImportError**: Se genera cuando Python no puede encontrar un módulo que estás intentando importar en tu código.


* 
**KeyError**: Se produce cuando intentas acceder a un campo o clave que no existe dentro de un diccionario.



---

## Excepciones de Entrada/Salida (I/O)

Son muy frecuentes al interactuar con el sistema de archivos o recursos externos.

* 
**FileNotFoundError**: Ocurre al intentar realizar operaciones como abrir, leer, escribir o borrar un archivo o directorio que no existe.


* 
**IOError**: Es una excepción más general relacionada con errores de entrada y salida, como intentar escribir en un archivo que no tiene permisos de escritura.



---

## Comparación Visual: Errores de Sintaxis vs. Excepciones

Es fundamental recordar que, aunque ambos detienen el programa si no se manejan, su naturaleza es distinta.

| Tipo | Momento de detección | Ejemplo |
| --- | --- | --- |
| **Error de Sintaxis** | Antes de que el programa se ejecute (fase de lectura).

 | Olvidar los dos puntos `:` al final de un `if`. |
| **Excepción** | Mientras el programa se está ejecutando (tiempo de ejecución).

 | Intentar dividir por cero durante un cálculo.

 |

El manejo de estas excepciones mediante sentencias `except` permite que el programa se recupere del error o notifique al usuario en lugar de bloquearse por completo.
