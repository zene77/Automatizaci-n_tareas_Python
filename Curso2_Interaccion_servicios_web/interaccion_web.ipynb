# Aplicaciones y servicios web

Una aplicación web es un programa con el que interactúas a través de HTTP, generalmente usando un navegador web.

Funcionamiento básico de una aplicación web
- El navegador envía una solicitud HTTP a un servidor web.
- El servidor web pasa la solicitud a la aplicación, que decide qué contenido mostrar y genera la página en formato HTML.

Contenido y datos en la aplicación web
- La aplicación también sirve imágenes y otros datos necesarios para que el navegador pueda mostrar la página correctamente.
- Muchas aplicaciones web ofrecen APIs que permiten interactuar con ellas mediante programas.

Interacción con APIs y servicios web
- Las APIs permiten enviar mensajes (llamadas API) a la aplicación sin usar la interfaz web tradicional.
- El programa que recibe estas llamadas se llama endpoint de API, y la comunicación se basa en protocolos que ambos, el servicio y el programa, entienden.

# API RESTful

Las APIs RESTful son una forma estándar y segura de permitir la comunicación entre aplicaciones a través del protocolo HTTP.

Conceptos clave de RESTful APIs
- Utilizan métodos HTTP como GET, POST, PUT y HEAD para realizar operaciones CRUD (crear, leer, actualizar, eliminar) sobre recursos.
- Cada solicitud recibe una respuesta con un código de estado HTTP, siendo 200 un código de éxito común.

Formato y seguridad
- JSON es el formato más común para intercambiar datos entre cliente y servidor en APIs RESTful, por ser legible y fácil de procesar.
- Las APIs RESTful añaden una capa de protección sobre recursos como bases de datos, usando autenticación con tokens o claves API para controlar el acceso.

Compatibilidad y uso
- Son compatibles con cualquier lenguaje de programación que soporte HTTP/HTTPS, como Python, JavaScript, Java, entre otros.
- Permiten separar la lógica de control de los datos, facilitando la implementación de funciones adicionales como limitación de uso, análisis y almacenamiento en caché.

# ¿Qué es la arquitectura REST?

La lectura explica la arquitectura REST, un estilo para diseñar aplicaciones y servicios web que permite la comunicación entre clientes y servidores de manera estándar y eficiente.

Principios clave de REST
- REST se basa en seis restricciones: interfaz uniforme, sin estado, cacheable, cliente-servidor, sistema en capas y código bajo demanda (opcional).
- Estas restricciones aseguran escalabilidad, simplicidad e interoperabilidad en sistemas RESTful.

Uso del protocolo HTTP con REST
- REST funciona sobre HTTP, usando métodos estándar como GET, POST, PUT, DELETE y PATCH para interactuar con recursos.
- Las solicitudes incluyen toda la información necesaria y pueden usar encabezados para autenticación o formatos de datos como JSON o XML.

Modelo de madurez Richardson
- Este modelo clasifica APIs REST en cuatro niveles según su cumplimiento con las restricciones REST, desde nivel 0 (básico) hasta nivel 3 (HATEOAS completo).
- HATEOAS permite que las respuestas del servidor incluyan enlaces para descubrir recursos relacionados sin codificar URIs en el cliente.

Resumen final
- REST es una arquitectura basada en principios claros que facilita la creación de APIs limpias y extensibles.
- Es importante diseñar bien las APIs desde el inicio, ya que son difíciles de cambiar una vez publicadas y en uso.

La arquitectura **REST** (Representational State Transfer) es, en esencia, un conjunto de reglas que permiten que dos computadoras se comuniquen a través de internet de forma eficiente y escalable. Si piensas en internet como una gran biblioteca, REST es el sistema de fichas y etiquetas que permite a cualquier usuario encontrar, pedir o modificar un libro sin importar de dónde venga.

Aquí te detallo los conceptos clave para terminar de aterrizar la información que compartiste:

---

### 1. El concepto de "Recurso"

En REST, todo es un **recurso**. Un usuario, una foto, un comentario o un producto en una tienda son recursos. Cada recurso tiene un identificador único llamado **URI** (Uniform Resource Identifier), que suele verse como una dirección web.

### 2. Los "Verbos" HTTP (Acciones)

Para que el cliente le diga al servidor qué hacer con esos recursos, se utilizan los métodos estándar de HTTP. Es como un lenguaje universal:

* **GET**: "Léeme este recurso" (Consultar).
* **POST**: "Crea un nuevo recurso" (Añadir).
* **PUT/PATCH**: "Actualiza este recurso" (Modificar).
* **DELETE**: "Elimina este recurso" (Borrar).

---

### 3. Las Restricciones Explicadas de forma Sencilla

Para que una API sea considerada realmente **RESTful**, debe cumplir con los principios que mencionaste. Los más importantes en el día a día son:

* **Sin estado (Stateless):** Imagina que vas a un restaurante y cada vez que pides algo, tienes que recordarle al camarero quién eres y en qué mesa estás. El servidor no "guarda" tu sesión; cada mensaje debe llevar toda la información necesaria para entenderse por sí solo.
* **Interfaz Uniforme:** No importa si pides una lista de usuarios o una lista de coches, la forma de pedirlo y la estructura de la respuesta (usualmente en **JSON**) deben ser consistentes.
* **Cliente-Servidor:** El que hace la interfaz (frontend) y el que guarda los datos (backend) están separados. Esto permite que puedas cambiar la aplicación del móvil sin tener que tocar la base de datos.

---

### 4. ¿Por qué es tan importante el Modelo de Madurez de Richardson?

Mucha gente dice que hace "REST", pero en realidad solo están enviando datos por HTTP. El modelo de Richardson es una regla de medir:

* **Nivel 0 y 1:** Son niveles básicos donde solo se usa la red para enviar mensajes.
* **Nivel 2:** Es el estándar de la industria. Usas los verbos correctamente (GET para leer, POST para crear) y códigos de estado (como el famoso **404** si algo no existe o **200** si todo va bien).
* **Nivel 3 (HATEOAS):** Es la "perfección". El servidor no solo te da el dato, sino que te dice qué puedes hacer después. Si pides un balance bancario, la respuesta te incluye enlaces para "ingresar dinero" o "transferir", de modo que el cliente no tiene que adivinar las URLs.

---

### Ejemplo Práctico

Si estuvieras diseñando una API para una tienda de libros, seguirías esta estructura:

1. **GET** `/libros` -> Devuelve la lista de todos los libros.
2. **GET** `/libros/42` -> Devuelve los detalles del libro con ID 42.
3. **POST** `/libros` -> (Con datos en el cuerpo) Crea un nuevo libro en el sistema.
4. **DELETE** `/libros/42` -> Borra ese libro específico.



# Uso de API REST para acceder a datos web

El contenido explica cómo acceder a datos web utilizando APIs RESTful, que permiten la comunicación entre clientes y servidores para obtener información.

Pasos clave para acceder a datos con APIs RESTful
- Identificar el endpoint de la API, que es la URL específica para el recurso deseado.
- Seleccionar el método HTTP adecuado según la acción: GET, POST, PUT, PATCH o DELETE.

Preparación y envío de la solicitud
- Configurar los encabezados necesarios, como tokens de autenticación y tipo de contenido.
- Preparar el cuerpo de la solicitud para métodos que envían datos (POST, PUT).
- Enviar la solicitud HTTP usando herramientas o librerías de programación.

Recepción y manejo de la respuesta
- Recibir la respuesta con código de estado, encabezados y cuerpo con los datos.
- Analizar la respuesta para extraer la información y manejar errores según el código recibido.

Consideraciones adicionales para el uso de APIs
- Implementar autenticación y autorización según lo requiera la API.
- Manejar paginación, filtrado, límites de tasa y errores para un uso eficiente y seguro.

Este contenido explica cómo se pueden usar frameworks y herramientas en Python para trabajar con APIs REST, tanto para consumirlas como para crear servicios propios.

Herramientas para consumir APIs REST en Python

- FastAPI es un framework moderno y rápido, con buena documentación, aunque con una comunidad más pequeña por ser reciente.

Buenas prácticas y recomendaciones
- Mantener consistencia en nombres de parámetros y uso de métodos HTTP para facilitar la colaboración.
- Documentar las APIs y fomentar la experimentación para crear un ecosistema sólido.
- Validar las respuestas de las APIs para manejar posibles discrepancias con la documentación.

---

## 1. Herramientas del Lado del Cliente (Consumo)

Estas bibliotecas se utilizan cuando tu programa necesita "hablar" con un servidor externo para obtener o enviar datos.

## 2. Herramientas del Lado del Servidor (Creación)

Marcos de trabajo (frameworks) para construir la infraestructura que recibirá las peticiones.

### Comparativa de Frameworks Populares

### **Flask** (Micro-framework)

Es como una caja de herramientas minimalista. Tú decides qué piezas añadir.

* **Ventajas:** Curva de aprendizaje suave, total libertad de estructura.
* **Desventajas:** No incluye validación de datos nativa ni soporte asíncrono robusto (en versiones antiguas).

### **Django / Django REST Framework (DRF)**

Es la opción "con todo incluido". Proporciona autenticación, base de datos y panel de administración por defecto.

* **Ventajas:** Muy seguro y escalable. Ideal para aplicaciones empresariales complejas.
* **Desventajas:** Es pesado y tiene una estructura rígida que puede ser difícil de aprender al principio.

### **FastAPI** (Alto Rendimiento)

El estándar moderno. Utiliza tipado de datos de Python (`pydantic`) para validar la información automáticamente.

* **Ventajas:** Documentación automática con Swagger (puedes probar la API en el navegador al instante) y velocidad comparable a Node.js o Go.
* **Desventajas:** Al ser más reciente, algunos paquetes antiguos de Python podrían no estar optimizados para su flujo asíncrono.

---

## 3. Mejores Prácticas de Diseño (Resumen)

Para que tu API sea profesional y fácil de usar por otros desarrolladores, sigue estos principios:

* **Nombres consistentes:** Usa `snake_case` (común en Python) o `camelCase`, pero no mezcles ambos.
* **Uso correcto de Verbos HTTP:**
* `GET`: Recuperar información.
* `POST`: Crear un nuevo recurso.
* `PUT`/`PATCH`: Actualizar datos existentes.
* `DELETE`: Eliminar un recurso.


* **Códigos de Estado:** No respondas siempre con un `200 OK`. Si algo no existe, usa `404`; si hay un error de permisos, `403`.
* **Validación:** Nunca confíes en los datos que envía el cliente. Valida siempre el formato y el contenido antes de procesarlos.

---

### ¿Cómo elegir?

* Si eres **principiante** y quieres aprender las bases: **Flask + Requests**.
* Si quieres construir algo **moderno y rápido**: **FastAPI + HTTPX**.
* Si necesitas una **aplicación robusta** con base de datos compleja: **Django + DRF**.

# ¿Qué es Flask?

## ¿Qué es Flask?

Flask es lo que en el mundo de la programación conocemos como un **microframework**. Está escrito en Python y su objetivo principal es permitirte crear aplicaciones web de forma increíblemente rápida y con el mínimo código posible.

A diferencia de otros marcos de trabajo más pesados, Flask no intenta adivinar qué necesitas. No te obliga a usar una base de datos específica ni te impone una estructura de carpetas rígida. Es como un lienzo en blanco que viene con las herramientas básicas necesarias para conectar tu código con la web.

---

### Características Principales

* **Es un Microframework:** No significa que Flask sea "pequeño" o incapaz de hacer cosas grandes, sino que su **núcleo es simple pero extensible**. Si necesitas algo (como validación de usuarios o conexión a bases de datos), simplemente instalas un "plugin" o extensión.
* **Basado en WSGI y Jinja2:**
* **WSGI (Werkzeug):** Es la pieza que permite que Flask se comunique con el servidor web.
* **Jinja2:** Es el motor de plantillas que permite generar páginas HTML dinámicas usando lógica de Python.


* **Libertad Total:** Tú decides cómo organizar tus archivos y qué librerías usar. Esto lo hace ideal tanto para principiantes que quieren entender cómo funciona una web, como para expertos que necesitan control total sobre la arquitectura.

---

### Flask vs. Django: La eterna comparación

Es común dudar entre estos dos. Aquí tienes la diferencia clave:

| Aspecto | Flask | Django |
| --- | --- | --- |
| **Filosofía** | "Hazlo tú mismo" (Minimalista). | "Baterías incluidas" (Todo viene listo). |
| **Curva de aprendizaje** | Muy rápida y sencilla. | Más lenta debido a su complejidad. |
| **Estructura** | Flexible y libre. | Rígida y estandarizada. |
| **Uso ideal** | Microservicios, prototipos, APIs. | Aplicaciones masivas y complejas. |

---

### ¿Quién usa Flask?

A pesar de ser un "micro" framework, su potencia es suficiente para soportar infraestructuras de nivel mundial. Algunas empresas que lo utilizan son:

* **Netflix:** Para muchas de sus herramientas internas de gestión.
* **Reddit:** Parte de su infraestructura funciona sobre Flask.
* **Uber:** Lo utiliza para microservicios específicos que requieren alta velocidad.
* **Airbnb:** Para diversas aplicaciones internas y herramientas de análisis de datos.

---

### Ejemplo rápido: Tu primera aplicación

Para que veas lo simple que es, este es el código necesario para crear un servidor web funcional que diga "Hola Mundo":

```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def hola_mundo():
    return '¡Hola! Este es un servidor Flask.'

if __name__ == '__main__':
    app.run()

```

Con solo esas líneas, Flask levanta un servidor local en tu computadora listo para recibir visitas. Es esta simplicidad lo que lo convierte en uno de los frameworks más queridos por la comunidad de Python.

# Serialización de datos

La **serialización de datos** es el puente que permite que la información "viva" de un programa (objetos, listas, diccionarios en la memoria RAM) pueda viajar a través de un cable o guardarse en un archivo sin perder su estructura.

Como bien mencionas, cuando pasamos de una estructura plana (como un CSV) a una estructura anidada o jerárquica, necesitamos formatos más potentes.

Aquí te presento los tres formatos de serialización más comunes que se utilizan hoy en día para manejar estos datos complejos:

---

## 1. JSON (JavaScript Object Notation)

Es el estándar de oro para las **APIs Web**. Aunque su nombre menciona JavaScript, es totalmente independiente del lenguaje y Python lo maneja de forma nativa con el módulo `json`.

* **Ventajas:** Muy ligero, fácil de leer para humanos y máquinas.
* **Estructura:** Se parece casi exactamente a los diccionarios y listas de Python que mostraste en tu ejemplo.
* **Uso ideal:** Comunicación entre un servidor y una aplicación móvil o web.

---

## 2. YAML (YAML Ain't Markup Language)

YAML es conocido por ser aún más legible que JSON porque elimina las llaves `{}` y los corchetes `[]`, utilizando la **sangría (indentación)** para marcar la jerarquía, igual que Python.

* **Ventajas:** Extremadamente limpio. Permite añadir comentarios dentro del archivo.
* **Estructura:**
```yaml
- name: Sabrina Green
  phone:
    office: 802-867-5309
    cell: 802-867-5310

```


* **Uso ideal:** Archivos de configuración de programas (como Docker o Kubernetes).

---

## 3. XML (Extensible Markup Language)

Es el "abuelo" de los formatos modernos. Utiliza etiquetas similares a HTML para definir los datos.

* **Ventajas:** Muy estricto y excelente para documentos que requieren una validación formal muy fuerte.
* **Estructura:**

```xml
    <person>
        <name>Sabrina Green</name>
        <phone>
            <office>802-867-5309</office>
        </phone>
    </person>
    ```
*   **Uso ideal:** Sistemas bancarios antiguos, servicios web SOAP y archivos de configuración de software corporativo.

# Cómo utilizar Flask

Utilizar **Flask** es entrar en un mundo donde tú tienes el control total de la arquitectura. A diferencia de otros frameworks que te dan una estructura rígida, Flask te entrega los cimientos y tú decides cómo construir el resto.

## 1. El Ciclo de Vida de una Petición

Para usar Flask, es vital entender qué sucede desde que alguien escribe una URL hasta que ve el contenido. Flask actúa como un intermediario que "escucha" rutas específicas.

### Estructura de un código básico:

1. **Importación:** `from flask import Flask`.

Como viste en el ejemplo, no quieres crear una ruta manual para cada usuario. Flask usa **parámetros de ruta** entre corchetes angulares:

```python
@app.route('/perfil/<username>')

    # Aquí podrías buscar a 'username' en una base de datos

```

# Formatos de serialización de datos

En el mundo del desarrollo, elegir el formato de serialización adecuado es como elegir el idioma en el que vas a escribir un mensaje: depende de quién sea el destinatario y para qué se use el mensaje.

Basado en la lectura, aquí tienes el análisis detallado de cómo trabajar con estos formatos en Python.

## 1. JSON (JavaScript Object Notation)

Es el formato más utilizado para la comunicación en la web (APIs REST).

### Cómo usarlo en Python:

El módulo nativo `json` de Python es la herramienta estándar. Como muestra el ejemplo, la función **`json.dump()`** se encarga de transformar un diccionario o lista de Python directamente en un archivo físico.

* **Parámetro `indent`:** Es vital para nosotros los humanos. En el código `json.dump(people, people_json, indent=2)`, el número indica cuántos espacios de sangría se aplicarán, haciendo que el archivo sea legible en lugar de una sola línea gigante de texto.
* **Dato técnico:** JSON solo soporta tipos de datos básicos (strings, números, booleanos, listas y diccionarios). No puede guardar directamente objetos complejos de Python como una conexión a una base de datos.

---

## 2. YAML (YAML Ain't Markup Language)

YAML destaca por su extrema limpieza visual. Es el favorito para archivos de configuración que los humanos deben editar frecuentemente.

### Cómo usarlo en Python:

A diferencia de JSON, YAML no suele venir preinstalado en Python (se usa comúnmente la librería `PyYAML`). La función **`yaml.safe_dump()`** es la recomendada por seguridad, ya que evita la ejecución de código malicioso que podría estar escondido en un archivo YAML desconocido.

* **Diferencia visual:** Mientras JSON usa llaves `{}` y comas `,`, YAML usa guiones `-` para listas e identación para la jerarquía.
* **Orden de los campos:** Notarás que en el ejemplo YAML, los campos aparecen en orden alfabético (`department`, `name`, `phone`). Esto es normal en la serialización de diccionarios, ya que lo importante es la relación clave-valor, no el orden.

---

## 3. Comparativa: ¿Cuándo usar cuál?

| Característica | **JSON** | **YAML** |
| --- | --- | --- |
| **Uso Principal** | Transmisión de datos entre servicios web (APIs). | Archivos de configuración de software. |
| **Legibilidad** | Buena, pero se ensucia con muchas llaves y comillas. | Excelente, parece una lista escrita a mano. |
| **Comentarios** | No permite comentarios (estándar oficial). | Sí permite comentarios usando el símbolo `#`. |
| **Popularidad** | Universal en casi cualquier lenguaje de programación. | Muy popular en herramientas de infraestructura (Docker, Ansible). |

## 4. Otros formatos mencionados

Aunque el curso se centra en JSON y YAML, existen alternativas para casos específicos:

* **Pickle (Python):** Solo para Python. Es muy rápido pero peligroso si abres archivos de fuentes desconocidas.
* **Protocol Buffers:** Creado por Google. Es un formato binario (no legible por humanos) diseñado para ser extremadamente rápido y pequeño.
* **XML:** El estándar antiguo. Es muy descriptivo y pesado, pero sigue siendo vital en sistemas bancarios y gubernamentales.

### Regla de oro:

Si estás enviando datos a una **API**, usa **JSON**. Si estás escribiendo los ajustes de configuración de tu **aplicación**, usa **YAML**.


**Pickle** es el sistema de serialización nativo y exclusivo de Python. Su función principal es transformar jerarquías de objetos de Python en un **flujo de bytes** (binario) para que puedan ser guardados en un archivo o enviados por una red, y luego reconstruidos exactamente como estaban.

A diferencia de JSON o YAML, que son formatos de texto plano, Pickle es un formato **binario**.

---

### ¿Para qué sirve realmente?

Su uso principal es la **persistencia del estado** de un programa. Aquí tienes los casos más comunes:

* **Guardar modelos de Machine Learning:** Es el estándar para guardar un modelo ya entrenado (con librerías como Scikit-learn). Entrenar un modelo puede tardar horas; con Pickle, lo guardas en un segundo y lo cargas mañana listo para usar.
* **Cache de objetos complejos:** Si tu programa realiza un cálculo matemático pesado que genera un objeto complejo, puedes "picklear" ese resultado. La próxima vez que abras el programa, cargas el objeto instantáneamente en lugar de recalcularlo.
* **Transferencia de datos en computación distribuida:** Cuando usas librerías como `multiprocessing` o `PySpark`, Python usa Pickle internamente para enviar objetos de un núcleo del procesador a otro.

---

### ¿Cómo funciona el proceso?

1. **Pickling (Serialización):** Convierte el objeto (una lista, un diccionario, o incluso una clase propia) en una serie de bytes.
2. **Unpickling (Deserialización):** Lee esos bytes y vuelve a crear el objeto original en la memoria RAM.

---

### La gran advertencia: ¿Por qué es "peligroso"?

La documentación oficial de Python lo dice claramente: *"Nunca abras un archivo pickle de una fuente en la que no confíes"*.

* **Ejecución de código arbitrario:** A diferencia de JSON, que solo guarda datos (texto/números), Pickle puede guardar **instrucciones de ejecución**. Un atacante puede enviarte un archivo `.pickle` malicioso que, al momento de ser "des-serializado" (`pickle.load()`), ejecute un comando para borrar tu disco duro o instalar un virus.
* **Incompatibilidad:** Como es exclusivo de Python, no puedes leer un archivo Pickle con Java, C++ o JavaScript. Además, a veces un archivo guardado con Python 2 no se puede leer bien en Python 3.

---

### Comparativa rápida

| Característica | JSON | Pickle |
| --- | --- | --- |
| **Formato** | Texto (legible) | Binario (no legible) |
| **Compatibilidad** | Universal (cualquier lenguaje) | Solo Python |
| **Seguridad** | Seguro (solo datos) | **Riesgoso** (puede ejecutar código) |
| **Capacidad** | Limitado a tipos básicos | Casi cualquier objeto de Python |

**En resumen:** Usa **JSON** si vas a compartir datos con el mundo o con otros lenguajes. Usa **Pickle** solo para guardar cosas internas de tu propio programa de Python que necesites recuperar rápidamente más tarde.

# 📝 **Resumen de las 10 preguntas del laboratorio**

1. **Procesar archivos en diccionarios antes de enviarlos**  
   → Permite **manipular y transformar los datos fácilmente** antes de enviarlos al servicio web.

2. **Usar JSON en lugar de pickle**  
   → JSON es **más seguro**, especialmente cuando los datos vienen de fuentes externas.

3. **Propósito de un marco web (framework)**  
   → Proporciona **estructura y componentes preconstruidos** para crear aplicaciones web.

4. **Propósito de la librería requests**  
   → Gestionar **peticiones HTTP** (GET, POST, etc.) hacia servicios web.

5. **Función de os.listdir()**  
   → **Lista archivos y directorios** dentro de una ruta especificada.

6. **Limitación de REST**  
   → REST es **sin estado (stateless)**: el servidor no guarda información de sesión del cliente.

7. **Propósito del método HTTP GET**  
   → **Solicitar datos** a un servidor.

8. **Qué hace un ORM**  
   → Mapea **objetos del lenguaje** a **tablas de base de datos**, simplificando el acceso.

9. **Comando para mostrar un archivo**  
   → `cat example.txt` muestra el contenido del archivo.

10. **Por qué comprobar status_code en un POST**  
   → Para saber si la **solicitud tuvo éxito**.
