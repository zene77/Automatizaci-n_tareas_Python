```python
#!/usr/bin/env python3

import os
from PIL import Image

origen = os.path.expanduser("~/images")
destino = "/opt/icons"

os.makedirs(destino, exist_ok=True)

for archivo in os.listdir(origen):

    # Ignorar archivos ocultos como .DS_Store
    if archivo.startswith("."):
        continue

    ruta_in = os.path.join(origen, archivo)

    img = Image.open(ruta_in)

    # Rotar 90° clockwise
    img = img.rotate(-90, expand=True)

    # Redimensionar a 128x128
    img = img.resize((128, 128))

    # Convertir a RGB
    img = img.convert("RGB")

    # Guardar como JPEG
    nombre = archivo + ".jpeg"
    ruta_out = os.path.join(destino, nombre)
    img.save(ruta_out, "JPEG")

print("Imágenes procesadas correctamente.")
```
