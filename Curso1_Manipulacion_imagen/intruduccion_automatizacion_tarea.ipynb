# Sistemas distribuidos

Los sistemas distribuidos son conjuntos de componentes de software que trabajan juntos aunque estén en servidores separados, con el objetivo de eliminar cuellos de botella y puntos únicos de falla.

Características principales de los sistemas distribuidos
- Compartición de recursos: pueden compartir hardware, software o datos.
- Procesamiento simultáneo y escalabilidad: múltiples máquinas pueden procesar la misma función y se puede aumentar la capacidad añadiendo más máquinas.

Ventajas de los sistemas distribuidos
- Flexibilidad: cada servidor puede ajustarse según el componente que aloje.
- Tolerancia a fallos y redundancia: si un nodo falla, otros pueden tomar su lugar para mantener el sistema operativo.

Desventajas de los sistemas distribuidos
- Complejidad aumentada: son más difíciles de diseñar y administrar que sistemas centralizados.
- Problemas de red: pueden introducir nuevos puntos de falla, retrasos y costos adicionales por la escalabilidad.

En resumen, los sistemas distribuidos son esenciales para aplicaciones que requieren alta disponibilidad y escalabilidad, pero requieren un diseño cuidadoso para manejar su complejidad y posibles problemas.

# NALSD

NALSD (Non-Abstract Large System Design) es un proceso desarrollado por Google para que los ingenieros de confiabilidad de sitios (SRE) puedan diseñar y evaluar sistemas grandes y complejos con un enfoque práctico y realista.

Fases del diseño NALSD
- Fase 1: Diseño técnico iterativo que incluye prototipos, estudios de viabilidad y retroalimentación para responder si el diseño es posible y cómo mejorarlo.
- Fase 2: Evaluación de la escalabilidad y resiliencia del sistema ante aumentos significativos de carga, asegurando que funcione bien a gran escala y sea rentable.

Objetivos clave de NALSD
- Planificación de capacidad adecuada para dimensionar recursos y prever crecimiento sin desperdicio ni falta de recursos.
- Aislamiento de componentes para mantener simplicidad, modularidad e independencia entre partes del sistema.
- Degradación gradual que permite que el sistema siga funcionando parcialmente cuando alguna parte falla, evitando fallos totales.

Recurso adicional
- Google ofrece un Workbook de NALSD con prácticas y guías para diseñar sistemas grandes y confiables, útil para profundizar en este tema.

# ¿Qué es una API?

Las Interfaces de Programación de Aplicaciones (APIs) facilitan la comunicación entre diferentes programas de software.

Definición y función de las APIs
- Las APIs son conjuntos de funciones, clases y métodos públicos que permiten a otros programas usar funcionalidades sin necesidad de reescribir código.
- Pueden ser utilizadas incluso por software desarrollado en diferentes lenguajes de programación, como en servicios en la nube que se comunican mediante llamadas web.

Diferencia entre funciones internas y APIs
- Las bibliotecas contienen funciones internas o privadas que no están destinadas a ser usadas directamente, sino que apoyan las funciones públicas que forman la API.
- La API actúa como una promesa de estabilidad: aunque el código interno cambie, la forma de llamar a las funciones y sus resultados deben mantenerse constantes.

Manejo de cambios en las APIs
- Los autores de bibliotecas pueden mejorar el código interno, pero no deben cambiar la interfaz pública para evitar romper el código que depende de ella.
- Cambios que rompen la compatibilidad deben comunicarse claramente y reservarse para incrementos mayores de versión. Familiarizarse con la API es esencial para usar correctamente una biblioteca.

# ¿Cómo dar sentido a una API?

Este contenido explica cómo aprender a usar una biblioteca o API nueva, enfocándose en la importancia de entender su organización, funciones y documentación.

Familiarización con la biblioteca o API
- Es normal que tome tiempo entender cómo funciona una biblioteca nueva, pero es valioso dedicar ese tiempo para comprender sus funciones, entradas y salidas.
- Una buena API debe ser descriptiva, con nombres de funciones claros que indiquen su propósito.

Uso de la documentación y ejemplos
- La documentación es clave para entender qué hace cada función; por ejemplo, el código de Pillow muestra cómo abrir, rotar y mostrar una imagen de forma intuitiva.
- En bibliotecas de código abierto, se puede revisar el código fuente para confirmar el comportamiento esperado; en APIs cerradas, se debe confiar en la documentación.

Docstrings y ayuda integrada en Python
- Las bibliotecas en Python suelen documentar sus funciones con docstrings, que son accesibles mediante la función help() o en el entorno de desarrollo.
- La documentación completa de Pillow está disponible en línea, incluyendo referencias y tutoriales para facilitar el aprendizaje.

# Generar y gestionar contenedores

El contenido explica cómo crear y gestionar imágenes de contenedores Docker para aplicaciones, destacando la importancia de construir imágenes personalizadas a partir de un Dockerfile.

Elección de la imagen base
- La imagen base proporciona el sistema operativo dentro del contenedor y debe ser ligera y adecuada para la aplicación.
- Ejemplos comunes incluyen Debian, Ubuntu, Alpine Linux y Python, cada uno con diferentes versiones y variantes.

Creación del Dockerfile
- El Dockerfile contiene instrucciones paso a paso para construir la imagen, como definir la imagen base, copiar archivos, instalar dependencias y exponer puertos.
- Un ejemplo muestra un Dockerfile para una aplicación Python con Flask y SQLAlchemy, explicando cada línea.

Construcción y gestión de imágenes
- Se construye la imagen con el comando docker build, etiquetándola para facilitar su identificación y posible subida a un registro.
- Se pueden listar, eliminar imágenes no usadas y ejecutar contenedores a partir de las imágenes creadas, gestionando así el espacio y versiones.

# Generar y gestionar contenedores: VS Studio y Docker

La integración de Visual Studio (VS) Code con Docker facilita la creación y gestión de contenedores directamente desde el entorno de desarrollo.

Integración de Docker en VS Code
- VS Code ofrece una extensión oficial de Docker que se instala desde el menú de extensiones.
- Esta extensión proporciona autocompletado, resaltado de sintaxis para Dockerfile, plantillas rápidas y una vista para gestionar contenedores e imágenes.

Funciones principales de la extensión Docker
- Permite iniciar, detener contenedores, abrir una terminal dentro de ellos y explorar sus archivos.
- Facilita la conexión con DockerHub y otros registros para publicar imágenes mediante arrastrar y soltar.

DevContainer y desarrollo en contenedores
- DevContainer es un estándar abierto que permite desarrollar y depurar código dentro de un contenedor sin reconstruirlo constantemente.
- Compatible con GitHub Codespaces, posibilita trabajar en entornos de desarrollo en la nube con configuración versionada.

Resumen clave
- La integración simplifica la creación, orquestación y gestión de contenedores, mejorando la eficiencia en desarrollo y despliegue.
- Los contenedores aseguran un entorno consistente y aislado, eliminando problemas de compatibilidad y facilitando la colaboración en equipo.
